```python
#defining shopify and os library
import os
import shopify

#token API that can be found in your shopify setting,
#make sure to allow access to your required data
#merchant name is found in your profile

token = 'TOKEN'
merchant = 'MERCHANT'
                     
api_session = shopify.Session(merchant, '2024-01', token)
                     
shopify.ShopifyResource.activate_session(api_session)

def get_data(object_name):
    all_data = []
    attribute = getattr(shopify, object_name)
    data = attribute.find(since_id=0, limit=250)
    for d in data:
        all_data.append(d)
    while data.has_next_page():
        data=data.next_page()
        for d in data:
            all_data.append(d)
    return all_data

orders = get_data('Product')
```
