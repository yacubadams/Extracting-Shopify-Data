```python
#defining shopify and os library
import os
import shopify

#token API that can be found in your shopify setting,
#make sure to allow access to your required data
#merchant name is found in your profile

token = 'token'
merchant = 'merchant_name'
                     
api_session = shopify.Session(merchant, '2024-01', token)
                     
shopify.ShopifyResource.activate_session(api_session)

def get_data(object_name):
    all_data = []
    attribute = getattr(shopify, object_name)
    data = attribute.find(since_id=0, limit=250)
    for d in data:
        all_data.append(d)
    while data.has_next_page():
        data=data.next_page()
        for d in data:
            all_data.append(d)
    return all_data

products = get_data('Product')
    
    
import pandas as pd

# Assuming products is a list of Shopify Product objects obtained using get_data('Product')
# Make sure to adjust the variable name accordingly

# Create a list of dictionaries containing product attributes
products_data = []
for product in products[0:100]:
    products_data.append(product.attributes)

# Create a Pandas DataFrame from the list of dictionaries
df = pd.DataFrame(products_data)

# Display the DataFrame
print(df)    

# Save the DataFrame to a CSV file
csv_file_path = "product_data.csv"
df.to_csv(csv_file_path, index=False)
print(f"Data saved to {csv_file_path}")

```
